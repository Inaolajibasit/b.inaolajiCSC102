```python
# project 3 
import pandas as pd
lager = ["33 Exports", "Desperados", "Goldberg", "Gulder", "Heineken", "Star"]
stout = ["Legend", "Turbo king", "Williams"]
non_alcoholic = ["Maltina", "Amstel Malt", "Malta Gold", "Fayrouz"]

class drinks():
    
    
    def read_lager(self, lager):
        print("Lager", lager)
        df = pd.DataFrame(lager)
        df.to_csv('lager.csv')
        df
    
    def read_stout(self, stout):
        print("Stout", stout)
        df = pd.DataFrame(stout)
        df.to_csv('stout.csv')
        df
        
    def read_non_alcoholic(self, non_alcoholic):
        print("non_alcoholic", non_alcoholic)
        df = pd.DataFrame(non_alcoholic)
        df.to_csv('non_alcoholic.csv')
        df
        
        
        
cup = drinks()
print("what category of drink do you want?")
print("1 for Lager ")
print("2 for Stout ")
print("3 for Non_Alcoholic ")
val = int(input("what do you want"))
if val == 1:
    print(" you chose 1")
    cup.read_lager(lager)
elif val == 2:
    print("you chose 2")
    cup.read_stout(stout)
elif val == 3:
    print("you chose 3")
    cup.read_non_alcoholic(non_alcoholic)
```

```python
#project 2
print("what do you want to calculate?")
print("0 for Area of Trapezium")
print("1 for Area of Rhombus")
print("2 for Area of Parallelogram")
print("3 for Volume of Cube")
print("4 for Volume of Cylinder")
val = int(input("Choose one: "))

class solve():
    
    def trapezium(self):
        print("to claculate the area of a Trapezium")
        self.h = int(input("input height = "))
        self.b1 = int(input("input base1 = "))
        self.b2 = int(input("input base2 = "))
        area = (self.h / 2) * (self.b1 + self.b2)
        print("Area is ",area)
        
        
    def rhombus(self):
        print("to claculate area of a Rhombus")
        self.d1 = int(input("input diagonal 1 = "))
        self.d2 = int(input("input diagonla 2 = "))
        area = 0.5 * (self.d1 * self.d2)
        print("Area is ",area)
        
    def parallelogram(self):
        print("to claculate area of a Parallelogram")
        self.b = int(input("input base = "))
        self.a = int(input("input altitude = "))
        area = self.a * self.b
        print("Area is ",area)

    def cube(self):
        print("to claculate Volume of a Cube")
        self.l = int(input("input length of side = "))
        area = 6 * (self.l * self.l )
        print("Area is ",area)
        
    def cylinder(self):
        print("to claculate area of a Cylinder")
        self.r = int(input("input radius = "))
        self.h = int(input("input height = "))
        area = 3.142 * self.r * self.r * self.h
        print("Area is ",area)
        
bobo = solve()
if val == 0:
    bobo.trapezium()
elif val == 1:
    print("yaY")
    bobo.rhombus()
elif val == 2:
    print("yaY")
    bobo.parallelogram()
elif val == 3:
    print("yaY")
    bobo.cube()
elif val == 4:
    print("yaY")
    bobo.cylinder()
```
